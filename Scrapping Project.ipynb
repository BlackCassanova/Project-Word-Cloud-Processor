### Load Page


```python
import requests
from bs4 import BeautifulSoup as bs
```

```python
#loading the Webpage
r = requests.get("https://keithgalli.github.io/web-scraping/webpage.html")

#converting to a beautiful soup
soup = bs(r.content)

#prin the html
print(soup.prettify())
```

### Starting Scrape 

```python
first_header = soup.find("h2")
first_header
```

```python
headers = soup.find_all("h2")
headers
```

```python
#Passing a list of elements to look for
first_header =soup.find(['h1','h2'])
first_header
```

```python
headers = soup.find_all(['h1', 'h2', 'h3'])
headers
```

```python
# Attribute Passer
para =  soup.find_all("p" , attrs = {"id":"footer"})
para
```

```python
body = soup.find('body')
div =  body.find('div')
img = div.find('img')
img
```

```python
# Searching for specific strings
import re
para = soup.find_all("p", string =  re.compile("minivan"))
para
```

### Get Text

```python
header = soup.find(['p'])
print(header.get_text())
```
