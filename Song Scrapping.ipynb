```python
import requests
from bs4 import BeautifulSoup as bs
import pickle
import time
```

```python
time.sleep(2)
r = requests.get("https://www.azlyrics.com/lyrics/bollywood/namonamo.html")
#converting to a beautiful soup
soup = bs(r.content)
#print the html
print(soup.prettify())
```

```python
song_name = soup.find("div", attrs = {"class": "ringtone"}).find_next_sibling()
print(song_name.get_text())
```

```python
Song_lyrics = soup.find("div", attrs = {"class": "ringtone"}).find_next_sibling("div")
print(Song_lyrics.get_text())
```

```python
def album_lyrics(text):
    time.sleep(7)
    r = requests.get(text).text
    soup = bs(r)
    content = soup.find("div", attrs = {"class": "ringtone"}).find_next_sibling("div")
    lyrics = content.get_text()
    return lyrics
```

```python
urls =["https://www.azlyrics.com/lyrics/bollywood/namonamo.html",
       "https://www.azlyrics.com/lyrics/bollywood/sweetheart.html",
       "https://www.azlyrics.com/lyrics/bollywood/qaafirana.html",
       "https://www.azlyrics.com/lyrics/bollywood/jaannisaar.html"]

songs = ["Namo Namo","Sweetheart","Qaafirana","Jaan Nisaar"]
```

```python
song_lyrics = [album_lyrics(u) for u in urls]       
```

```python
!mkdir Kedarnath
```

```python
for i,c in enumerate(songs):
    with open ("Kedarnath/" + c + ".txt", "wb") as file:
        pickle.dump(song_lyrics[i], file)
```

```python
import warnings
warnings.filterwarnings('ignore')
import pickle
```

```python
data = {}
for i, c in enumerate (songs):
    with open("Kedarnath/" + c + ".txt", "rb") as file:
        data[c] = pickle.load(file)
```

```python
data.keys()
```

```python
#Converting the Dictionary to a Pandas Dataframe
import pandas as pd

pd.set_option('max_colwidth', 200)

df1 = pd.DataFrame.from_dict(data, orient='index')
df1 = df1.reset_index()
df1.columns = ['Songs', 'Lyrics']
df1
```

```python
df2 = df1.copy()
df2['Album'] = 'Kedarnath'
df2
```

```python
df3 = df2.groupby(['Album'])['Lyrics'].apply(' '.join).reset_index()
df3
```
